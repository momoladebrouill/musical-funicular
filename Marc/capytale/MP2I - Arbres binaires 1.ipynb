# Premières fonctions

On va considérer le type des arbres binaires `'a arbre` défini par :

````
type 'a arbre_bin = 
    | Nil 
    | Noeud of 'a arbre_bin * 'a * 'a arbre_bin
````

**Question 1**

Écrire des fonctions permettant de calculer la taille et la hauteur d’un
arbre, ainsi qu’une fonction permettant de déterminer si un arbre est
réduit à une feuille.

<div class="ui styled accordion"> <div class="title">     <i class="dropdown icon"></i>
Réponse </div> <div class="content">

````
let rec taille (a : 'a arbre_bin) : int = 
    match a with
    Nil -> 0
    | Noeud (g,x,d) -> max (taille g) (taille d) + 1
````

````
assert (taille Nil = 0);
assert (taille (Noeud(Nil,3,Nil)) = 1);
assert (taille (Noeud(Nil,'a',Nil)) = 1);
assert (taille (Noeud(Nil,'a',Noeud(Nil,'b',Nil))) = 2);
"Valide"
````

````
let rec hauteur (a : 'a arbre_bin) : int =
    match a with
    Nil -> -1
    | Noeud (g,x,d) -> max (hauteur g) (hauteur d) + 1
````

````
assert (hauteur Nil = -1);
assert (hauteur (Noeud(Nil,3,Nil)) = 0);
assert (hauteur (Noeud(Nil,'a',Nil)) = 0);
assert (hauteur (Noeud(Nil,'a',Noeud(Nil,'b',Nil))) = 1);
"Valide"
````

````
let feuille (a : 'a arbre_bin) : bool =
    match a with
    Noeud (Nil,a,Nil) -> true
    | _ -> false
    
````

````
assert (not (feuille Nil));
assert (feuille (Noeud(Nil,3,Nil)));
assert (feuille (Noeud(Nil,'a',Nil)));
assert (not (feuille (Noeud(Nil,'a',Noeud(Nil,'b',Nil)))));
"Valide"
````

**Question 2**

Dans cette question, on va identifier un nœud à son étiquette.

Déterminer des fonctions renvoyant, sous forme d’une liste d’étiquettes,
les nœuds, les feuilles et les nœuds internes d’un arbre.

Allez regarder les tests si vous avez un doute sur ce qui doit être
renvoyé.

<div class="ui styled accordion"> <div class="title">     <i class="dropdown icon"></i>
Réponse </div> <div class="content">

````
let rec noeuds (a : 'a arbre_bin) : 'a list = 
    let rec aux a acc =
    match a with
    | Nil -> acc
    | Noeud (g,x,d) -> x::(noeuds g)@(noeuds d)
    in aux a []
````

````
assert(noeuds Nil = []);
assert(noeuds (Noeud(Nil,1,Nil)) = [1]);
let a = Noeud(
    Noeud(Nil, 2, Noeud(Nil, 3, Nil)),
    4, Noeud(Noeud(Nil, 5, Nil), 6,
        Noeud(Nil, 7, Nil))) in
assert(List.sort Stdlib.compare (noeuds a)
    = [2;3;4;5;6;7]);
"Valide"
````

````
let rec feuilles (a : 'a arbre_bin) : 'a list = 
    let rec aux a acc =
        match a with
        | Nil -> acc
        | Noeud (g,x,d) when feuille a -> x::acc
        | Noeud (g,x,d) -> (aux g acc)@(aux d acc)
    in aux a []
````

````
assert(feuilles Nil = []);
assert(feuilles (Noeud(Nil,1,Nil)) = [1]);
let a = Noeud(
    Noeud(Nil, 2, Noeud(Nil, 3, Nil)),
    4, Noeud(Noeud(Nil, 5, Nil), 6,
        Noeud(Nil, 7, Nil))) in
assert(List.sort Stdlib.compare (feuilles a)
= [3;5;7]);
"Valide"
````

````
let rec noeuds_internes (a : 'a arbre_bin) : 'a list = 
    let rec aux a acc =
        match a with
        | Nil -> acc
        | Noeud (g,x,d) when feuille g && feuille d -> x::acc
        | Noeud (g,x,d) when not (feuille a) -> x::(aux g acc)@(aux d acc)
        | Noeud (g,x,d) -> (aux g acc) @ (aux g acc)
        
    in aux a []
````

````
assert(noeuds_internes Nil = []);
assert(noeuds_internes (Noeud(Nil,1,Nil)) = []);
let a = Noeud(
    Noeud(Nil, 2, Noeud(Nil, 3, Nil)),
    4, Noeud(Noeud(Nil, 5, Nil), 6,
        Noeud(Nil, 7, Nil))) in
assert(List.sort Stdlib.compare
    (noeuds_internes a)
= [2;4;6]);
"Valide"
````

**Question 3**

Écrire une fonction permettant de déterminer si un arbre binaire est
strict **et non vide**.

<div class="ui styled accordion"> <div class="title">     <i class="dropdown icon"></i>
Réponse </div> <div class="content">

````
let rec est_strict (a : 'a arbre_bin) : bool =
    match a with 
    Nil -> false
    | Noeud (Nil,x,Nil) -> true
    | Noeud (g,x,d) when Nil = g || Nil = d -> false
    | Noeud (g,x,d) -> (est_strict g) && (est_strict d)
````

````
assert(not (est_strict Nil));
assert(est_strict (Noeud(Nil, 1, Nil)));
let a = Noeud(
    Noeud(Nil, 2, Noeud(Nil, 3, Nil)),
    4, Noeud(Noeud(Nil, 5, Nil), 6,
        Noeud(Nil, 7, Nil))) in
assert(not (est_strict a));
let b = Noeud(
    Noeud(Nil, 2, Nil),
    4, Noeud(Noeud(Nil, 5, Nil), 6,
        Noeud(Nil, 7, Nil))) in
assert(est_strict b);
"Valide"
````

**Question 4**

Écrire une fonction qui renvoie les sous-arbres, en dupliquant les
sous-arbres identiques, d’un arbre.

<div class="ui styled accordion"> <div class="title">     <i class="dropdown icon"></i>
Réponse </div> <div class="content">

````
let rec sous_arbres (a : 'a arbre_bin) : 'a arbre_bin list = 
    let rec aux a acc =
        match a with
        Nil -> [a]
        | Noeud (g,_,d) -> a::(aux g acc)@(aux d acc)
    in aux a []
````

````
assert(sous_arbres Nil = [ Nil ]);
let a = Noeud(Nil, 1, Nil) in 

assert(List.sort Stdlib.compare (sous_arbres a)
    = [ Nil; Nil; a ]);
let b = Noeud(Nil, 2, Nil) in
let c = Noeud(a, 3, b) in 
assert(List.sort Stdlib.compare
    (sous_arbres c) =
    [Nil; Nil; Nil; Nil; a; b; c]);
"Valide"
````

Pour accéder à un nœud de l’arbre, on descend en partant de la racine et
en allant à gauche ou à droite. On peut donc représenter un tel chemin
par une liste de déplacements :

````
type deplacement = Gauche | Droite
type chemin = deplacement list
````

**Question 5**

Écrire une fonction `chemin_noeud : 'a arbre_bin -> chemin -> 'a option`
qui renvoie l’étiquette d’un nœud donné par son chemin. On utilise un
type `option` en cas de chemin invalide.

<div class="ui styled accordion"> <div class="title">     <i class="dropdown icon"></i>
Réponse </div> <div class="content">

````
let rec chemin_noeud (a : 'a arbre_bin) (l : chemin) : 'a option =
    match a with
    Nil -> None
    | Noeud (g,x,d) ->
        match l with
        [] -> Some x
        | Droite::q -> chemin_noeud  d q
        | Gauche::q -> chemin_noeud g q
````

````
let a = Noeud(
    Noeud(Nil, 2, Noeud(Nil, 3, Nil)),
    4, Noeud(Noeud(Nil, 5, Nil), 6,
        Noeud(Nil, 7, Nil))) in
assert(chemin_noeud a [Gauche; Gauche] = None);
assert(chemin_noeud a [] = Some 4);
assert(chemin_noeud a [Droite; Gauche] = Some 5);
assert(chemin_noeud a [Droite; Droite] = Some 7);
assert(chemin_noeud a [Droite; Droite; Droite] = None);
"Valide"
````

**Question 8**

Écrire une fonction d’évaluation qui calcule le résultat de cette
expression.

On pourra utiliser `int_of_string` pour convertir une chaine de
caractère comme `"42"` en l’entier `42`.

<div class="ui styled accordion"> <div class="title">     <i class="dropdown icon"></i>
Réponse </div> <div class="content">

````
let rec evalue a = 
    match a with
    Nil -> 0
    | Noeud (g,x,d) -> 
        match x with 
        "+" -> evalue g + evalue d
        |"*" -> evalue g * evalue d
        |"/" -> evalue g / evalue d
        | _ -> int_of_string x
````

````
let e = Noeud(Noeud( Noeud(Nil, "2", Nil), "+", Noeud(Nil, "3", Nil)), "*", Noeud(Nil, "5", Nil)) in
assert(evalue e = 25);
"Valide"
````

# Arbres binaires stricts

Pour représenter des arbres dont les nœuds ont tous deux fils non vides,
on va utiliser le type

````
type ('a, 'b) arbre_strict = 
    Feuille of 'a 
    | NoeudI of ('a, 'b) arbre_strict * 'b * ('a, 'b) arbre_strict
````

Ce type permet de représenter naturellement des expressions
arithmétiques comme :

````
type operations = Plus | Mult
````

````
let e = NoeudI(
    NoeudI(Feuille 2, Plus, Feuille 3), 
    Mult, 
    Feuille 5)
````

**Question 9**

Écrire une fonction `evalue : (int, operations) arbre_strict -> int` qui
évalue une telle expression.

<div class="ui styled accordion"> <div class="title">     <i class="dropdown icon"></i>
Réponse </div> <div class="content">

````
let rec evalue a = 
    match a with 
    Feuille x -> x
    |NoeudI (g,x,d) -> 
        match x with 
        Plus -> evalue g + evalue d
        | Mult -> evalue g * evalue d
````

````
let e = NoeudI(NoeudI( Feuille 2, Plus, Feuille 3), Mult, Feuille 5) in
assert(evalue e = 25);
"Valide"
````


````
let rec arbre_strict_vers_arbre_bin a =
    let asvab = arbre_strict_vers_arbre_bin in
    match a with
    | Feuille x -> Noeud(Nil, F x, Nil) 
    | NoeudI (g,x,d) -> Noeud(asvab g,N x,asvab d)
````

````
let asvab = arbre_strict_vers_arbre_bin in
assert(asvab (Feuille 1) = Noeud(Nil, F 1, Nil));
assert(asvab (NoeudI(Feuille 'a', 1, NoeudI(Feuille 'b', 2, Feuille 'c')))
    = Noeud(Noeud(Nil,F 'a',Nil), N 1, Noeud(Noeud(Nil,F 'b',Nil), N 2, Noeud(Nil, F 'c', Nil))));
"Valide"
````
